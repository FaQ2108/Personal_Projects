```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

print("Bibliotecas importadas y configuradas correctamente.")
```

```python
# Lista de empresas (Ejemplo: Bancario - JPM, BAC; Tecnológico - AAPL, MSFT)
empresas = ['JPM', 'BAC', 'AAPL', 'MSFT']

# Descarg de datos, de Yahoo Finance
datos = yf.download(empresas, start="2018-01-01", end="2023-01-01")

# Visualizacion de los primeros datos descargados
print(datos.head())
```

### Verificación de Integridad de Datos

Una parte fundamental de cualquier análisis de datos es asegurar la integridad y calidad de los mismos. Para ello, hemos verificado si existen datos faltantes en nuestro dataset de precios de cierre ajustados para las empresas seleccionadas.

```python
# Seleccion de las columnas de precios de cierre ajustados
datos_cierre = datos['Adj Close']

# Verificacion presencia o ausencia de datos faltantes
datos_faltantes = datos_cierre.isnull().sum()
print("Datos faltantes en cada columna:\n", datos_faltantes)

# Visualizacion de las primeras filas de los datos de cierre ajustados
print(datos_cierre.head())
```

Este resultado indica que nuestros datos están completos y listos para ser utilizados en el análisis exploratorio, asegurando la fiabilidad de los insights que se obtendrán a partir de aquí.

## Gráfico de Líneas de los Precios de Cierre Ajustados
Este gráfico muestra la evolución de los precios de cierre ajustados a lo largo del tiempo para cada empresa.

```python
# Gráfico de líneas de los precios de cierre ajustados
plt.figure(figsize=(14, 7))
for empresa in empresas:
    plt.plot(datos_cierre.index, datos_cierre[empresa], label=empresa)

plt.title('Precios de Cierre Ajustados de Acciones')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre Ajustado')
plt.legend()
plt.show()
```

* AAPL (Apple): La línea verde también muestra un crecimiento robusto, aunque un poco menos volátil que Microsoft. Apple ha mantenido un crecimiento sostenido gracias a sus innovaciones en hardware y servicios.

* JPM (JPMorgan Chase & Co.): La línea azul presenta una tendencia al alza más moderada con algunas fluctuaciones. La estabilidad relativa puede estar ligada a la naturaleza del sector bancario y su respuesta a las políticas económicas.

* BAC (Bank of America): La línea naranja muestra la menor volatilidad y crecimiento, sugiriendo una estabilidad típica del sector bancario, aunque con menos crecimiento comparado con las tecnológicas.

###Comparaciones Entre Sectores:

  - Tecnología vs Bancario: Las empresas tecnológicas (AAPL y MSFT) han experimentado un mayor crecimiento y volatilidad en comparación con las empresas bancarias (JPM y BAC). Esto es consistente con las tendencias de mercado donde las tecnológicas han visto un auge significativo en valor de mercado.

  - Volatilidad: Microsoft y Apple, siendo líderes tecnológicos, muestran mayores fluctuaciones, lo que puede ser resultado de sus constantes innovaciones y respuestas del mercado a sus productos. En contraste, JPMorgan y Bank of America muestran estabilidad propia del sector financiero.

  - Eventos Clave:
Los picos y valles en las líneas podrían corresponder a eventos específicos de la empresa o del mercado (lanzamientos de productos, informes de ganancias, cambios en políticas económicas, etc.). Analizar estos eventos puede proporcionar un contexto adicional para las fluctuaciones.

Conclusión:
Este gráfico nos proporciona una visión clara de cómo se han comportado estas acciones a lo largo de los últimos cinco años. ## Gráfico de Cajas (Box Plot)

  - IQR: De 150 a 250 USD.

  - Valores Atípicos: Algunos valores fuera del IQR, reflejando una mayor volatilidad y potencial de crecimiento en el mercado tecnológico.

###Interpretación:

## Análisis de Correlación
Analizamos la correlación entre los precios de cierre ajustados de las diferentes empresas para identificar relaciones y patrones.

```python
# Matriz de correlación
correlacion = datos_cierre.corr()
print("Matriz de correlación de precios de cierre ajustados:\n", correlacion)

# Heatmap de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(correlacion, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación de Precios de Acciones')
plt.show()
```

###Observaciones:
* AAPL (Apple) y MSFT (Microsoft)

  - Correlación de 0.973: Esto sugiere una relación muy fuerte entre las acciones de Apple y Microsoft, lo cual tiene sentido dado que ambas son grandes tecnológicas que probablemente respondan a factores económicos y del mercado de manera similar.

* BAC (Bank of America) y JPM (JPMorgan Chase)

  - Correlación de 0.946: Existe una fuerte correlación entre estas dos acciones del sector bancario, lo cual es esperado ya que ambas operan bajo regulaciones similares y se ven afectadas por políticas económicas y tasas de interés.

* Intersector (AAPL y BAC, AAPL y JPM, MSFT y BAC, MSFT y JPM)

  - Correlaciones de 0.735 a 0.783: Las correlaciones entre las tecnológicas (AAPL, MSFT) y las bancarias (BAC, JPM) son moderadas. Esto sugiere que, aunque hay cierta relación, las acciones de diferentes sectores no siempre se mueven en perfecta sincronía. Esto puede deberse a que diferentes factores impactan a los sectores tecnológicos y financieros.

* BAC (Bank of America) y MSFT (Microsoft)

  - Correlación de 0.774: Aunque es menor que la correlación dentro de los mismos sectores, sigue siendo significativa. Esto podría reflejar respuestas comunes a eventos macroeconómicos.

###Conclusión:
* Fuerte Correlación Dentro de Sectores: Las acciones dentro del mismo sector (tecnología o bancario) tienden a moverse de manera similar, probablemente debido a influencias macroeconómicas comunes y a tendencias sectoriales.

* Moderada Correlación Entre Sectores: Las acciones de diferentes sectores muestran una correlación menos fuerte, indicando que los factores específicos del sector juegan un rol importante en sus movimientos.

Este análisis de correlación nos ayuda a entender mejor cómo las acciones pueden moverse en conjunto o de manera independiente, proporcionando insights valiosos para estrategias de diversificación y gestión de portafolios.
